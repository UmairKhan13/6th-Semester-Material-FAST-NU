```python
def get_max_int_number(chrmosome_length):
    return((2**chrmosome_length)-1)



def initial_population(population_size,chrmosome_length):
    import random
    random_numbers=[]
    for i in range(population_size):
            random_numbers.append(str(bin(random.randint(0,get_max_int_number(chrmosome_length)))[2:]).zfill(chrmosome_length)) 
    return(random_numbers)



def fitness_function(random_numbers):
    fitness_result=[]
    for i in range (0,len(random_numbers)):
        fitness_result.append(int(random_numbers[i],2))        
    return(fitness_result)



def selection(random_numbers,fitness_result):
    min_index=fitness_result.index(min(fitness_result))
    max_index=fitness_result.index(max(fitness_result))
    random_numbers[min_index]=random_numbers[max_index]    
    return(random_numbers)


    

def cross_over(random_numbers):
    for i in range(0,len(random_numbers)-1,2):
        temp_1=(random_numbers[i])[:4]+(random_numbers[i+1])[4:]
        temp_2=(random_numbers[i+1])[:4]+(random_numbers[i])[4:]
        random_numbers[i]=temp_1
        random_numbers[i+1]=temp_2
    return(random_numbers)

      
    
def mutation(random_numbers):
    for i in range (0,len(random_numbers)):
        random_numbers[i]=random_numbers[i].replace('0','1',1)
    return(random_numbers)
    
def Max_condition_not_true(random_numbers,chrmosome_length):
    fitness_result=fitness_function(random_numbers)
    if(get_max_int_number(chrmosome_length) in fitness_result):
        return(False)
        
    return(True)
    

def main():
    
    population_size=4
    chrmosome_length=20
    random_numbers=initial_population(population_size,chrmosome_length)
    print("Initial random numbers are ",random_numbers,"\n\n\n\n")
    fitness_result=[]
    generations=-1
    while(Max_condition_not_true(random_numbers,chrmosome_length)):
        print("Numbers for iteration ",generations+1," are : ",random_numbers)
        fitness_result=fitness_function(random_numbers)
        print("Fitness function of iteration ",generations+1," are : ",fitness_result)
    
        random_numbers=selection(random_numbers,fitness_result)
        print("Numbers after selection are : ",random_numbers)

        random_numbers=cross_over(random_numbers)
        print("Numbers after crossover are : ",random_numbers)

        random_numbers=mutation(random_numbers)
        print("Numbers after mutation are : ",random_numbers)
        
        generations=generations+1
        print("\n\n\n\n")
    print("The number of generations are : ",generations)

        
    
main()

```
